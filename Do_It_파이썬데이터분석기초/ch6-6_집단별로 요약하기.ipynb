## 집단별로 요약하기

```python
import pandas as pd
exam = pd.read_csv('exam.csv')
```

```python
# math 평균 구하기
exam.agg(mean_math = ('math', 'mean'))
```

```python
# nclass별로 분리하기
# math 평균 구하기
exam.groupby('nclass')\
    .agg(mean_math = ('math', 'mean'))
```

```python
# 변수를 인덱스로 바꾸지 않기
exam.groupby('nclass', as_index = False)\
    .agg(mean_math = ('math', 'mean'))
```

```python
# nclass별로 분리
# 수학 점수 평균, 수학 점수 합계, 수학 점수 중앙값, 빈도(학생 수)
exam.groupby('nclass')\
    .agg(mean_math   = ('math', 'mean'),
         sum_math    = ('math', 'sum'),
         median_math = ('math', 'median'),
         n           = ('nclass', 'count'))
```

```python
# 모든 변수의 요약 통계량 한 번에 구하기
exam.groupby('nclass').mean()
```

## 집단별로 다시 집단 나누기

```python
mpg = pd.read_csv('mpg.csv')

# 제조 회사 및 구동 방식별 분리
# cty 평균 구하기
mpg.groupby(['manufacturer', 'drv'])\
   .agg(mean_cty = ('cty', 'mean'))
```

```python
# audi의 drv별 빈도
mpg.query('manufacturer=="audi"')\
   .groupby(['drv'])\
   .agg(n = ('drv', 'count'))
```

```python
# chevrolet의 drv별 빈도
mpg.query('manufacturer == "chevrolet"')\
   .groupby(['drv'])\
   .agg(n = ('drv', 'count'))
```

```python
mpg['drv'].value_counts()
```

```python
# drv 빈도 구하기
# 데이터 프레임으로 바꾸기, 변수명 n으로 바꾸기
# n을 100을 초과한 경우 추출
mpg['drv'].value_counts()\
          .to_frame('n')\
          .query('n > 100')
```

## pandas 함수 조합하기

```python
mpg.query('category == "suv"')\
   .assign(total = (mpg['hwy'] + mpg['cty'])/2)\
   .groupby('manufacturer')\
   .agg(mean_tot = ('total', 'mean'))\
   .sort_values('mean_tot', ascending = False)\
   .head()
```
